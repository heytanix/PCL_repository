```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
# Placeholder paths to replace
SOURCE_DIR = 'path/to/PCL_repository/dataset_yolo/images/train'  # Replace with the full path to your 'train' folder (containing non-violent and violent subfolders)
OUTPUT_DIR = 'path/to/save/split_data'                           # Replace with the path where you want to save the split data (e.g., a new folder like cnn_splits)
```

```python
# Classes (based on your structure)
CLASSES = ['non-violent', 'violent']  # Adjust if your folder names differ (note: use exact folder names)
```

```python
# Split ratios (train + test should sum to 1.0; add validation if needed)
TRAIN_RATIO = 0.8  # 80% for training
TEST_RATIO = 0.2   # 20% for testing
```

```python
# Function to split and copy images
def split_data(source_dir, output_dir, classes, train_ratio, test_ratio):
    for cls in classes:
        cls_dir = os.path.join(source_dir, cls)
        if not os.path.exists(cls_dir):
            print(f"Warning: Directory {cls_dir} does not exist. Skipping class '{cls}'.")
            continue
        
        # Get list of image files
        images = [f for f in os.listdir(cls_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        if not images:
            print(f"Warning: No images found in {cls_dir}. Skipping class '{cls}'.")
            continue
        
        # Split the images
        train_imgs, test_imgs = train_test_split(images, test_size=test_ratio, random_state=42)
        
        # Copy to output directories
        for split, img_list in zip(['train', 'test'], [train_imgs, test_imgs]):
            dst_dir = os.path.join(output_dir, split, cls)
            os.makedirs(dst_dir, exist_ok=True)
            for img in img_list:
                src_path = os.path.join(cls_dir, img)
                dst_path = os.path.join(dst_dir, img)
                shutil.copy(src_path, dst_path)
    
    print(f'Data split complete! Train/Test ratio: {train_ratio}/{test_ratio}. Output saved to {output_dir}')

# Run the split
split_data(SOURCE_DIR, OUTPUT_DIR, CLASSES, TRAIN_RATIO, TEST_RATIO)
```
